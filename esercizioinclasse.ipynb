```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
###
##  X label name = Month Number
##  Y label name = Total profit
##
df =pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/fav/company_sales_data.csv')
```

```python

df.head(20)
```

```python

import calendar

for month_idx in range(1, 13):
    print (calendar.month_name[month_idx])
```

```python
month_word = {
1:'January',
2:'February',
3:'March',
4:'April',
5:'May',
6:'June',
7:'July',
8:'August',
9:'September',
10:'October',
11:'November',
12:'December'
}
```

```python
df['month_number'] = df['month_number'].map(month_word)
df
```

```python
df.rename(columns={'month_number':'month'},inplace=True)
df
```

```python
plt.figure(figsize=(12,10))
ax = plt.gca()
df.plot(x ='month',y='facecream',kind = 'line', ax=ax,color='red')
df.plot(x ='month',y='facewash',kind = 'line', ax=ax,color='blue')
df.plot(x ='month',y='facecream',kind = 'line', ax=ax,color='yellow')
df.plot(x ='month',y='toothpaste',kind = 'line', ax=ax,color='green')
df.plot(x ='month',y='bathingsoap',kind = 'line', ax=ax,color='m')
df.plot(x ='month',y='shampoo',kind = 'line', ax=ax,color='orange')
df.plot(x ='month',y='moisturizer',kind = 'line', ax=ax,color='c')
#plt.show()
```
