```python
def batman(x):
    return np.abs(x/7) * np.sqrt(3 - (np.abs(x)-4)**2 / 16) + (4 - np.abs(x)) * (1 - np.sqrt(np.abs(x)/2)) * np.sqrt(np.abs(np.abs(x)-4)/(np.abs(x)-4)) + 5 * np.sin(np.pi * np.abs(x)/2)

x = np.linspace(-7, 7, 1000)
y = batman(x)

plt.plot(x, y)
plt.show()
```

```python
url_superficie = "https://it.wikipedia.org/wiki/Primi_100_comuni_italiani_per_superficie"
```

```python
comuni_superficie = pd.read_html(url_superficie,
                                 attrs={"class":"wikitable"},
                                 header=0)
comuni_superficie = comuni_superficie[0]

#comuni_superficie.columns=[['Pos.','Comune','Regione','Provincia','Superficie']]
```

```python
comuni_superficie = comuni_superficie[1:]
```

```python
comuni_superficie['Superficie (km²)'] = comuni_superficie['Superficie (km²)'].astype('Int64')
#siccome c'è l'errore della virgola nella prima riga sotto la sup di roma, pandas considerava tutta la colonna come stringa per fixare,
#sta roba si è detto di iniziare dalla riga successiva e trasformare tutta la colonna in un integer (int64)
```

```python
comuni_superficie.groupby('Regione').mean() #funziona come il rapporto one to many di excel, raggruppa tutti i dati sotto le varie regioni
# e li mette insieme
```

```python
reg = comuni_superficie.groupby('Regione')
reg.aggregate([np.mean, np.std, pd.Series.count])
```

```python
comuni_superficie.groupby('Regione')['Superficie (km²)'].count() #raggruppo per grandezza
```

```python
comuni_superficie.groupby('Regione')['Superficie (km²)'].count().sort_values(ascending=False) 
#metto in ordine dal più grande al più piccolo
```
