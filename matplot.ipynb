```python
import matplotlib.pyplot as plt #importare le librerie che mi servono si fa così: con import, il nome della lib e il nome che gli do io
import numpy as np
from numpy.random import rand
```

```python
#vado a generare delle variabili con valori che sono degli array perché sto importando da numpy 
x = rand (100) #saranno 100 valori con compresi tra 0 e 1 ovviamente il 100 è un numero che decido io che è variabile
x
```

```python
y = rand(100) #se gli elementi tra x e y sono di numero diverso va in errore
y
```

```python
plt.scatter(x,y) #questo comando mi genera uno scatter plot tra gli array generati
#plt.show()
```

```python
x = rand(1000)
y = rand(1000)
plt.scatter(x,y,color='red') #mettendo color='nomecolore'mi permette di cambiare il colore degli elementi visualizzati in grafico
```

```python
plt.plot(x,y) #di default fa quello di linea in maniera consecutiva
```

```python
plt.plot([1,2,3,4],'-r') #'-r' è il comando che mi trasforma il plot in line plot
```

```python
#MATPLOT LIB È UNA LIBRERIA MODULARE! AGGIUNGO IN MANIERA ADDITIVA I MODULI CHE VOGLIO RAPPRESENTARE.
# Aggiungi i Label = i "nomi" degli elementi.
plt.plot([5,2,9,4], marker='o', linewidth=3, color='green')
plt.xlabel('label x')
#plt.show()
```

```python
#inserire due plot insieme
plt.plot([1,2,3,4], [1,4,9,16],)
plt.plot([1,3,5,8], [4,3,10,12],'--') #'--' fa la linea tratteggiata.
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],'ob')
plt.axis([0, 6, 0, 20]) #questo comando mi definisce i valori degli assi, in questo caso quello delle x da 0 a 6 e quello delle y da 0 a 20
#plt.show()
```

```python
x = np.linspace(0,10,100) #ci garantisce di creare un dominio con tutte le coppie di x che mi fanno creare un grafico accurato in quel range
y = np.exp(-x) #ho importato da numpy (np) la funzione esponenziale. 
plt.plot(x,y)
#plt.show()
```

```python
x = np.linspace(0, 10, 100) #il valore dello spessore e distanza delle linee visualizzate nel grafico

plt.figure()
plt.plot(x, np.sin(x), '-') #ho richiamato da numpy (np) le funzioni seno e coseno per il valore di y
plt.plot(x, np.cos(x), '--')
```

```python
#per ingrandire l'immagine e vederla meglio:
x = np.linspace(0, 10, 100)
y= np.sin(x)
z= np.cos(x)

plt.figure(figsize=(18,10)) #questo è il comando che deve essere sempre posto prima che mi permette di ingrandire il grafico senza
#cambiare scala

plt.plot(x,y, '-')
plt.plot(x,z, '--')
#se salvo l'immagine la salva come file .png
```
